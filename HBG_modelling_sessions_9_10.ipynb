```python
# ============================================================
# SESSION 9 & 10 — PANDAS & DATA ANALYSIS
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ============================================================
# PART 1 — PANDAS BASICS
# ============================================================

# Creating a Pandas Series
numbers = pd.Series([5, 10, 15, 20, 25])

print("Pandas Series")
print(numbers)

# Creating a DataFrame
student_data = {
    "Name": ["Anna", "David", "Lucas", "Maria"],
    "Age": [21, 22, 20, 23],
    "Grade": [88, 91, 85, 95]
}

students = pd.DataFrame(student_data)

print("\nDataFrame")
print(students)

# ============================================================
# PART 2 — SELECTING DATA
# ============================================================

# Selecting columns
print("\nNames column:")
print(students["Name"])

# Selecting rows with loc
print("\nFirst row using loc:")
print(students.loc[0])

# Selecting rows with iloc
print("\nSecond row using iloc:")
print(students.iloc[1])

# Boolean filtering
high_grades = students[students["Grade"] > 89]

print("\nStudents with grades above 89:")
print(high_grades)

# ============================================================
# PART 3 — BASIC VISUALIZATION
# ============================================================

# Line chart
students.plot(
    x="Name",
    y="Grade",
    kind="line",
    title="Student Grades"
)

plt.ylabel("Grade")
plt.show()

# Bar chart
students.plot(
    x="Name",
    y="Grade",
    kind="bar",
    title="Grades Bar Chart"
)

plt.ylabel("Grade")
plt.show()

# ============================================================
# PART 4 — TIME SERIES DATA
# ============================================================

# Creating dates
dates = pd.date_range(start="2024-01-01", periods=30)

# Creating random values
values = np.random.randint(50, 150, size=30)

# Creating DataFrame
sales_data = pd.DataFrame({
    "Date": dates,
    "Sales": values
})

# Convert to datetime
sales_data["Date"] = pd.to_datetime(sales_data["Date"])

# Set index
sales_data.set_index("Date", inplace=True)

print("\nSales Data")
print(sales_data.head())

# ============================================================
# PART 5 — TIME SERIES VISUALIZATION
# ============================================================

sales_data.plot(
    figsize=(10, 5),
    title="Daily Sales"
)

plt.ylabel("Sales")
plt.show()

# ============================================================
# PART 6 — MOVING AVERAGE
# ============================================================

sales_data["Moving_Average"] = (
    sales_data["Sales"]
    .rolling(window=5)
    .mean()
)

print("\nSales Data with Moving Average")
print(sales_data.head(10))

sales_data.plot(
    figsize=(10, 5),
    title="Sales and Moving Average"
)

plt.show()

# ============================================================
# PART 7 — SUMMARY STATISTICS
# ============================================================

print("\nSummary Statistics")
print(sales_data.describe())

# ============================================================
# PART 8 — FILTERING DATES
# ============================================================

filtered_sales = sales_data[
    sales_data.index >= "2024-01-15"
]

print("\nFiltered Sales Data")
print(filtered_sales)

# ============================================================
# PART 9 — RANDOM DATA ANALYSIS
# ============================================================

random_values = np.random.normal(
    loc=100,
    scale=15,
    size=200
)

random_series = pd.Series(random_values)

print("\nRandom Data Statistics")
print(random_series.describe())

# Histogram
random_series.plot(
    kind="hist",
    bins=20,
    title="Distribution of Random Values"
)

plt.xlabel("Value")
plt.show()
```

## Explanation of the Code/notes

This notebook focuses on Pandas and data analysis techniques in Python.

The first part of the code introduces Pandas Series and DataFrames. A Series represents a single column of data, while a DataFrame is a table containing rows and columns. These structures are commonly used for organizing and analyzing datasets.

The notebook also demonstrates different ways to select and filter data. The `.loc[]` function selects data using labels, while `.iloc[]` selects data using numerical positions. Boolean filtering is used to display only observations that satisfy a condition.

The code then creates visualizations using Matplotlib and Pandas plotting tools. Line charts and bar charts are used to represent trends and compare values visually.

The next section introduces time series data. A time series is a dataset where observations are ordered over time. Dates are converted into datetime format and used as the index of the DataFrame.

The notebook also calculates a moving average. Moving averages smooth short-term fluctuations and help identify long-term trends in the data.

Summary statistics are generated using the `.describe()` function. This provides measures such as the mean, standard deviation, minimum, and maximum values.

The notebook also demonstrates date filtering, where only observations from a specific period are selected.

Finally, NumPy is used to generate random data for statistical analysis. A histogram is plotted to visualize the distribution of the generated values.
