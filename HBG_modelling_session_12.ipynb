```python
# ============================================================
# SESSION 12 — REGRESSION & FINANCIAL DATA ANALYSIS
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
import yfinance as yf

# ============================================================
# PART 1 — SYNTHETIC LINEAR DATA
# ============================================================

# Generate synthetic data
np.random.seed(42)

x = np.linspace(0, 10, 100)
noise = np.random.normal(0, 4, 100)

y = 4 * x + 5 + noise

# Plot synthetic data
plt.figure(figsize=(10, 5))
plt.scatter(x, y)

plt.title("Synthetic Linear Data")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

# ============================================================
# PART 2 — LINEAR REGRESSION USING SKLEARN
# ============================================================

# Reshape x for sklearn
X = x.reshape(-1, 1)

# Create model
linear_model = LinearRegression()

# Train model
linear_model.fit(X, y)

# Predictions
predicted_values = linear_model.predict(X)

# Coefficients
slope = linear_model.coef_[0]
intercept = linear_model.intercept_

print("Linear Regression Results")
print("-------------------------")
print("Slope:", slope)
print("Intercept:", intercept)

# Plot regression line
plt.figure(figsize=(10, 5))

plt.scatter(x, y, label="Observed data")
plt.plot(x, predicted_values, color="red", label="Regression line")

plt.title("Linear Regression")
plt.xlabel("x")
plt.ylabel("y")

plt.legend()
plt.show()

# ============================================================
# PART 3 — HOUSE PRICE EXAMPLE
# ============================================================

# Example housing dataset
housing_data = pd.DataFrame({
    "Size": [50, 80, 100, 120, 150, 180, 200],
    "Price": [70000, 95000, 120000, 140000, 170000, 210000, 250000]
})

# Independent and dependent variables
X_house = housing_data[["Size"]]
y_house = housing_data["Price"]

# Train regression model
house_model = LinearRegression()
house_model.fit(X_house, y_house)

# Predict house price for 160 square meters
predicted_house_price = house_model.predict([[160]])

print("\nPredicted House Price")
print("---------------------")
print(predicted_house_price[0])

# Plot housing regression
plt.figure(figsize=(10, 5))

plt.scatter(
    housing_data["Size"],
    housing_data["Price"],
    label="House data"
)

plt.plot(
    housing_data["Size"],
    house_model.predict(X_house),
    color="green",
    label="Regression fit"
)

plt.xlabel("House Size")
plt.ylabel("Price")
plt.title("House Prices vs Size")

plt.legend()
plt.show()

# ============================================================
# PART 4 — GRADIENT DESCENT
# ============================================================

def gradient_descent(x, y, learning_rate=0.01, epochs=100):

    m = 0
    b = 0

    n = len(x)

    error_history = []

    for epoch in range(epochs):

        y_estimated = m * x + b

        error = np.mean((y - y_estimated) ** 2)
        error_history.append(error)

        derivative_m = (-2 / n) * np.sum(x * (y - y_estimated))
        derivative_b = (-2 / n) * np.sum(y - y_estimated)

        m = m - learning_rate * derivative_m
        b = b - learning_rate * derivative_b

    return m, b, error_history

# Apply gradient descent
m_final, b_final, history = gradient_descent(
    x,
    y,
    learning_rate=0.01,
    epochs=100
)

print("\nGradient Descent Results")
print("------------------------")
print("Final slope:", m_final)
print("Final intercept:", b_final)

# Plot cost reduction
plt.figure(figsize=(10, 5))

plt.plot(history)

plt.title("Cost Reduction During Gradient Descent")
plt.xlabel("Epoch")
plt.ylabel("Mean Squared Error")

plt.show()

# ============================================================
# PART 5 — FINANCIAL DATA ANALYSIS
# ============================================================

# Download stock data
start_date = "2025-01-01"
end_date = "2025-12-31"

apple = yf.download("AAPL", start=start_date, end=end_date)
market = yf.download("^GSPC", start=start_date, end=end_date)

# Select closing prices
apple_close = apple["Close"]
market_close = market["Close"]

# Calculate daily returns
apple_returns = apple_close.pct_change().dropna()
market_returns = market_close.pct_change().dropna()

# Convert to arrays
X_market = market_returns.values.reshape(-1, 1)
y_apple = apple_returns.values

# Train regression model
beta_model = LinearRegression()
beta_model.fit(X_market, y_apple)

# Beta coefficient
beta_value = beta_model.coef_[0]

print("\nFinancial Beta")
print("----------------")
print("Apple Beta:", beta_value)

# Predictions
market_prediction = beta_model.predict(X_market)

# Plot relationship
plt.figure(figsize=(10, 5))

plt.scatter(
    market_returns,
    apple_returns,
    alpha=0.5,
    label="Daily returns"
)

plt.plot(
    market_returns,
    market_prediction,
    color="red",
    label="Regression line"
)

plt.xlabel("Market Returns")
plt.ylabel("Apple Returns")

plt.title("Apple Returns vs Market Returns")

plt.legend()
plt.show()

# ============================================================
# PART 6 — DATAFRAME INDEXING
# ============================================================

random_df = pd.DataFrame(
    np.random.randn(4, 4),
    columns=["A", "B", "C", "D"]
)

print("\nRandom DataFrame")
print(random_df)

# Accessing values
print("\nColumn A:")
print(random_df["A"])

print("\nUsing loc:")
print(random_df.loc[0, "A"])

print("\nUsing iloc:")
print(random_df.iloc[0, 0])
```
